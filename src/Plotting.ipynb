```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
dirs = os.listdir('results/us/predictions/')

new_dirs = []
for i in dirs:
    i_split = i.split('-')
    if i_split[0] == 'LSTM':
        if i_split[-1] == '0.csv':
            if i_split[2] == '14':
                new_dirs.append(i)

for i in new_dirs:
    print(i)
        
```

```python
i=3
print('results/us/predictions/' + new_dirs[i])
data = pd.read_csv('results/us/predictions/' + new_dirs[i], parse_dates=True, index_col=0)
plt.plot(data.index, data['True'], color='black', label='True')
plt.plot(data.index, data['Pred'], color='red', label='mean')

try:
    plt.fill_between(data.index, 
                     data['Pred']-data['Std'],
                     data['Pred']+data['Std'],
                     color='red', 
                     alpha=0.5,
                     label='True')
except:
    pass
plt.show()
```

```python
for i in range(4):
    plt.subplot(4,1,i+1)
    data = pd.read_csv('results/us/predictions/' + new_dirs[i], parse_dates=True, index_col=0)
    plt.plot(data.index, data['True'], color='black', label='True')
    plt.plot(data.index, data['Pred'], color='red', label='mean')
    try:
        plt.fill_between(data.index, 
                         data['Pred']-data['Std'],
                         data['Pred']+data['Std'],
                         color='red',
                         alpha=0.5,
                         label='True')
    except:
        pass
plt.show()

```
